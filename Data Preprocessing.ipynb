```python
# Data Manipulation:

''' Q 1 Create a new column - 'total_benefitted' that is a sum of inmates benefitted through all modes. '''

prisoners['total_benefited']=prisoners.iloc[:,2:].sum(axis=1)
prisoners
```

```python
'''Q 2 Create a new row = 'totals' that is a sum of inmates benefitted through each mode across all states.  '''

prisoners_total=prisoners.append(prisoners.iloc[:,2:].sum(numeric_only=True),ignore_index=True)
prisoners_total
```

```python
# Plotting:

import numpy as np
import matplotlib.pyplot as plt
xlabels=prisoners['STATE/UT'].values
plt.figure(figsize=(20,10))
plt.xticks(np.arange(xlabels.shape[0]),xlabels,rotation='vertical',fontsize=18)
plt.xticks
plt.bar(np.arange(prisoners.values.shape[0]),prisoners['total_benefited'],align='edge')
plt.show()
```

```python
''' Make a pie chart that depits the ratio among different modes of benefits.'''

labels=prisoners.columns[2:-1]
sizes=prisoners_total.iloc[-1][2:-1].values
fig1,ax1=plt.subplots()
ax1.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax1.axis('equal') 
plt.show() 
```
